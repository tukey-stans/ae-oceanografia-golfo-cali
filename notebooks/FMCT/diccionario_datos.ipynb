## Coordenadas zona costera occidental GC

```python
file_name='Coordenadas zona costera occidental GC.csv'
df=pd.read_csv(f'{DATA_PATH}/raw/{file_name}')
df.info()
```

```python
df.columns
```

```python
df['latitude']=df['23.8']
df['longitude']=df['-109.3']
```

```python
new_row = pd.DataFrame([{'latitude': 23.8, 'longitude': -109.3}],index=[0])
df = pd.concat([df.iloc[:0], new_row, df.iloc[0:]]).reset_index(drop=True)
df
```

```python
df.drop(['-109.3','23.8','Unnamed: 1'],axis=1,inplace=True)
df
```

```python
df = df.reset_index(drop=True)      
df.index = range(1, len(df) + 1)
df.index.name = 'Estación muestral'
df
```

```python
columns=['Estación muestral','latitude','longitude']

descriptions=[
    'Numero identificador que indica la estación de cual se extraen las observaciones',
    'Coordenada de latitud (valores positivos significan Norte y negativos Sur)',
    'Coordenada de longitud (valores positivos significan Este y negativos Oeste)'
                  ]

types=[
    'int64',
    'float64',
    'float64'
]


data_dict=pd.DataFrame({'column':columns,'description':descriptions,'type':types})
data_dict
```
